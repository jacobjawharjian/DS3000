(5 pts) Extract the poultry consumption data, from 1994 to 2014 (inclusive), for South Africa, that is measured in thousand tonnes of carcass weight**. Note: the filtered dataframe should only contain the time and the values.

```python
safrica_df = meat_df.loc[meat_df['LOCATION'].isin(['ZAF'])]
zafpoult_df = safrica_df.loc[safrica_df['SUBJECT'].isin(['POULTRY'])]
zafpoult_df = zafpoult_df.loc[zafpoult_df['MEASURE'].isin(['THND_TONNE'])]
zafpoult_df = zafpoult_df[4:25]
zafpoult_df = zafpoult_df[['TIME','Value']]
zafpoult_df
```

The above statistics measure 21 values in the selected data set. The mean of the set is 1,224,045 tons of poultry. Standard deviation is 466,216 tons of poultry. The median of the set is 1,084,411 tons of poultry. 

(10 pts) Visualize the extracted data, using a line chart, and comment on the trend.

```python
zafpoult_df.plot(x='TIME', y='Value')
plt.xlim([1993, 2016])
plt.title('Poultry Consumption (Thnd Tons) in South Africa by Year, 1994 - 2014')
plt.xlabel('Year')
plt.ylabel('Poultry Consumed in Thousands of Tons')
```

The above visualization sees a somewhat constant climb in the thousands of tonnage of poultry consumed in South Africa on a yearly basis. 

(15 pts) Predict the poultry consumption for 2015 and 2016, using simple linear regression. Ensure that you visualize the original data and also include the predictions from the model.

```python
X = zafpoult_df['TIME'].to_numpy()
X = X.reshape(-1,1)

Y = zafpoult_df['Value'].to_numpy()
Y = Y.reshape(-1,1)

lr_model = LinearRegression(fit_intercept=True)
lr_model = lr_model.fit(X,Y)

print('The intercept is: ', lr_model.intercept_[0])
print('The slope is: ', lr_model.coef_[0,0])
```

```python
m = lr_model.coef_[0,0]
b = lr_model.intercept_[0]

fifteen_prediction = m * 2015 + b
sixteen_prediction = m * 2016 + b
```

```python
zafpoult_df.plot(x='TIME', y='Value')
plt.xlim([1993, 2017])
plt.ylim([0, 2200])
plt.title('Poultry Consumption (Thnd Tons) in South Africa by Year, 1994 - 2014')
plt.xlabel('Year')
plt.ylabel('Poultry Consumed in Thousands of Tons')

sns.regplot(data=zafpoult_df, x ='TIME', y='Value')

plt.plot(2015, fifteen_prediction, marker="o", markersize=6, markeredgecolor="green", markerfacecolor="green")
plt.plot(2016, sixteen_prediction, marker="o", markersize=6, markeredgecolor="green", markerfacecolor="green")


print('The prediction for poultry consumption (in thousands of tons) in 2015 is', fifteen_prediction)
print('The prediction for poultry consumption (in thousands of tons) in 2016 is', sixteen_prediction)
```

(5 pts) Calculate the R-squared and explain what this tells you about the model.

```python
print('R-Squared: ', lr_model.score(X,Y))
```

This R-Squared score suggests that the regression model is a very strong fit for the data, and thus a good model to draw predictions from.
