```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.model_selection import GridSearchCV

```

(5 pts) Perform any necessary data preparation steps and EDA.

```python

wine     = load_wine()

data1 = pd.DataFrame(data= np.c_[wine['data'], wine['target']],
                     columns= wine['feature_names'] + ['target'])



data1
```

(5 pts) Partition the data into train/test sets.

```python
X = data1[['ash', 'alcalinity_of_ash', 'magnesium', 'total_phenols', 'flavanoids', 'proanthocyanins']]
y = data1['target']

X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   test_size=0.3,
                                                   random_state=7,
                                                   stratify=y
                                                   )
```

(10 pts) Normalize the numeric features and encode any categorical features. Ensure that you use sklearn to normalize/encode the training data, after which transform the test set (there is an example in the lecture).

```python
scaler = StandardScaler()
scaler.fit(X_train[['ash', 'alcalinity_of_ash', 'magnesium', 'total_phenols', 'flavanoids', 'proanthocyanins']])

X_train[['ash', 'alcalinity_of_ash', 'magnesium', 'total_phenols', 'flavanoids', 'proanthocyanins']] = scaler.transform(X_train[['ash', 'alcalinity_of_ash', 'magnesium', 'total_phenols', 'flavanoids', 'proanthocyanins']])

```

(20 pts) Build the model that predicts the target variable, and use GridSearchCV to tune at least two hyperparameters, and perform cross validation. You'll need to read the documentation of the SVM algorithm to identify the hyperparameters.

```python
params = {'C': np.arange(1, 10), 
         'gamma': np.arange(1, 5)}

svm = SVC()
svm_grid = GridSearchCV(svm, params, cv=5)
svm_grid.fit(X_train, y_train)


```

(5 pts) Display: 1) the best model/parameters and 2) the best score that was obtained.

```python
print(svm_grid.best_params_)
print(svm_grid.best_score_)
```

(5 pts) How do you feel about the model? Did it accurately predict the target? Evaluate the scores from GridSearchCV for the training set and the test set to diagnose any bias-variance problems.

```python
X_test[['ash', 'alcalinity_of_ash', 'magnesium', 'total_phenols', 'flavanoids', 'proanthocyanins']] = scaler.transform(X_test[['ash', 'alcalinity_of_ash', 'magnesium', 'total_phenols', 'flavanoids', 'proanthocyanins']])

y_pred = svm_grid.predict(X_test)

print(y_pred)

print(svm_grid.score(X_test, y_test))
```

The model did not do well at all. This was a surprise because the GridSearchCV performed pretty strongly on the training set. I would imagine that with regard to bias-variance, this is an issue of variance as a result of overfit on the training model.
